```python
tweets = tweepy.Cursor(api.search, q="graphics card", lang='en', since="2020-01-01",tweet_mode='extended').items(2000)
tweets = [x.full_text for x in tweets]
tweets = pd.DataFrame(tweets, columns=['Tweets'])
```

```python
print(tweets)
```

```python
def cleanData(twt):
    twt = re.sub('#','',twt)
    twt = re.sub('\\n', ' ', twt) #remove endline character
    twt = re.sub('https?:\/\/\S+', '', twt)
    twt = re.sub('@', '', twt)
    return twt
```

```python
tweets["cleanTweet"] = tweets['Tweets'].apply(cleanData)
```

```python
tweets
```

```python
def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

def getSentiment(score):
    if score > 0:
        return "Positive"
    elif score == 0:
        return "Neutral"
    else:
        return "Negative"

tweets['Subjectivity'] = tweets['cleanTweet'].apply(getSubjectivity)
tweets['Polarity'] = tweets['cleanTweet'].apply(getPolarity)
tweets['Sentiment'] = tweets['Polarity'].apply(getSentiment)
```
